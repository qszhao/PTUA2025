```python
import random
import numpy as np
```

```python
def monte_carlo_pi(convergence_criteria):
    n = 0 
    d = 0 
    while True:
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1.0:
            n += 1
        d += 1
        pi_estimate = 4 * n / d
        if abs(pi_estimate - np.pi) / np.pi <= convergence_criteria:
            return d 
```

```python
criteria_list = [0.01, 0.001, 0.0001, 0.00001]
results = {}  
for criteria in criteria_list:
    draws_required = []
    for _ in range(10):  
        draws = monte_carlo_pi(criteria)
        draws_required.append(draws)
    results[criteria] = draws_required
```

```python
for criteria, draws in results.items():
    average_draws = np.mean(draws) 
    std_dev_draws = np.std(draws) 
    print(f"Convergence criteria: {criteria}")
    print(f"  Average number of draws: {average_draws}")
    print(f"  Standard deviation: {std_dev_draws}")
    print()
```
