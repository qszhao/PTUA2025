```python
import random
import numpy as np

pi = np.pi
```

```python
# Convergence criteria array
converges = [0.01, 0.001, 0.0001, 0.00001]

for i in converges:
    print(i)
    pi_s = []
    ds = []
    
    for sim in range(10):
        n = 0
        d = 0
        ratios = []
        xs = []
        ys = []
        
        while True:
            x = random.random()
            y = random.random()
            
            xs.append(x)
            ys.append(y)
            
            if x**2 + y**2 <= 1.0:
                n += 1
            
            d += 1
            
            ratio = 4 * n / d
            
            ratios.append(ratio)
            
            if abs(ratio - pi) / pi <= i:
                pi_s.append(ratio)
                ds.append(d)
                print(i)
                print(pi_s)
                print(ds)
                print(f"Average Number of draws {np.mean(ds)} with convergence criterion {i}")
                print(f"Average standard deviation of number of draws is  {np.std(ds)} with convergence criterion {i}")
                break
```
